```python
import pandas as pd
```

```python
df = pd.read_json("../database/transcripts.json")
df
```

```python
df["text"] = "Channel: " + df["channel_username"] + " Title: " + df["title"]
df["text"]
```

```python
df[df["transcript"].isna()]
```

```python
def get_text(line):
    return " ".join([json["text"] for json in line])

df["text"] += df["transcript"].apply(get_text)
```

```python
# Get the values of the "text" column and join them with "\n\n"
text_content = "\n\n".join(df["text"].values)

# Write the text content to a file
with open("../documents/transcript.txt", "w") as f:
    f.write(text_content)
```

```python
from sentence_transformers import SentenceTransformer, util

# Load pre-trained BERT model
model = SentenceTransformer('paraphrase-distilroberta-base-v1')
```

```python
# Compute embeddings
embeddings = model.encode(df["text"], convert_to_tensor=True)

# Compute embedding for "goal"
goal_embedding = model.encode(["goal"], convert_to_tensor=True)

# Find the closest sentence to "goal"
closest_index = util.pytorch_cos_sim(goal_embedding, embeddings).argmax().item()
closest_sentence = df["text"].iloc[closest_index]

print("Closest sentence to 'goal':", closest_sentence)
```
